ESP32（如NodeMCU-32S開發板）是一款高集成度、超低功耗的雙核Wi-Fi + 藍牙 SoC 晶片，廣泛應用於物聯網、自動化、機器人與嵌入式系統專案。下方整理了其核心特性、引腳功能及主要應用建議。

### 主要特點與規格
- 雙核 32-bit LX6 處理器，最高主頻 240MHz
- 內建Wi-Fi（802.11 b/g/n）與藍牙4.2（BLE和BR/EDR）
- SRAM容量320KB，Flash容量依PCB設計選配（常見4MB/8MB）
- 多達34個可程控GPIO腳位，支援ADC、DAC、I2C、SPI、UART、PWM等多種通訊協定
- 支援觸控感應、硬體PWM、模擬輸入/輸出、極低功耗模式

### Pinout配置與常用功能
- ADC通道：ADC0（GPIO36）、ADC1（GPIO39）、ADC2（GPIO34）等
- DAC：DAC1（GPIO25）、DAC2（GPIO26）
- I2C：I2C SCL（GPIO22）、I2C SDA（GPIO21）
- UART：TX0/RX0、TX2/RX2 等支援多組串口
- SPI/VSPI/HSPI：可選用多組腳位搭配
- GPIO可自定義，常見數位/模擬I/O接口皆有標示
- 3.3V與Vin腳提供外部電源選擇，GND腳多組
- 支援多組觸控感應功能（TOUCH0~TOUCH9）

### 開發特性與應用
- 支援Arduino、ESP-IDF、MicroPython、PlatformIO等開發環境，上手快速
- 適合製作智能家居、遠端遙控、數據收集、無線傳感器、網路機器人等應用
- 可直接驅動LED、步進馬達、繼電器模組，也常與感測器等配合
- 具備高擴展性與無線通訊能力，適合需快速原型實現物聯網智慧裝置的場景

ESP32腳位配置靈活，具豐富外設可選，適合高效開發各類無線連網自動化系統。使用時應注意不同GPIO腳位的預設/特殊功能，避免與外部模組衝突，確保穩定運行。

AS5600 感測器是一款磁性感測旋轉編碼器模組，提供高解析度角度量測，適合自動化、機器人與馬達控制應用。以下針腳說明與功能整理：

### 主要針腳功能
- **VCC**：模組電源（3.3V）
- **GND**：地線
- **OUT**：PWM 或類比電壓輸出角度訊號
- **SCL**：I2C 通訊時鐘（Serial Clock Line）
- **SDA**：I2C 通訊資料線（Serial Data Line）
- **GPO**：模式選擇腳（內部上拉為預設模式，接地為程式設計模式B）
- **DIR**：旋轉方向設置腳（接地時角度值隨順時針遞增，接 VCC 反方向遞增）

### 特色與應用
- 解析度高，單圈磁性感測精度可達 12 bit
- 可同時支援 PWM、I2C 與電壓類比三種角度輸出
- I2C 通訊可進行數位角度讀取，便於單片機資訊整合
- 適用於馬達角度偵測、舵機定位、旋轉控制裝置、機械臂關節等場合

### 典型接線方式
- **I2C連接**：VCC → 3.3V、GND → GND、SCL → 控制器時鐘腳（如ESP32的GPIO22）、SDA → 控制器數據腳（如ESP32的GPIO21）
- **PWM或類比輸出**：OUT 腳輸出併至 MCU ADC 端
- 依需要設置 GPO、DIR 腳，預設可不接

AS5600用於磁環與軸旋轉角度偵測，結構簡單、信號豐富，適合與ESP32、Arduino等控制器直接配合應用。
